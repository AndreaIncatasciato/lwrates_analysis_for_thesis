```python
lwtransitions_allnew_data = lwrad.H2.get_lw_transitions(excited_states_to_use='LW', lw_transitions_reference='U_19+S_15')
lwtransitions_allnew_df = pd.DataFrame.from_dict(lwtransitions_allnew_data)
lwtransitions_allnew_df.head()
```

```python
print(np.unique(lwtransitions_allnew_df['VL']))
print(np.unique(lwtransitions_allnew_df['JL']))
print(np.unique(lwtransitions_allnew_df['VU']))
print(np.unique(lwtransitions_allnew_df['JU']))
```

```python
lwtransitions_abgrall_data = lwrad.H2.get_lw_transitions(excited_states_to_use='LW', lw_transitions_reference='A_94')
lwtransitions_abgrall_df = pd.DataFrame.from_dict(lwtransitions_abgrall_data)
lwtransitions_abgrall_df.head()
```

```python
fig, ax = plt.subplots()
wl_min = np.minimum(lwtransitions_abgrall_df['wl'].min(), lwtransitions_abgrall_df['wl'].min())
wl_max = np.maximum(lwtransitions_abgrall_df['wl'].max(), lwtransitions_abgrall_df['wl'].max())

ax.hist(lwtransitions_allnew_df['wl'],
        weights=lwtransitions_allnew_df['frac_diss']*lwtransitions_allnew_df['f'],
        bins=100, range=(wl_min, wl_max), histtype='step', log=True)
ax.hist(lwtransitions_u19_df['wl'],
        weights=lwtransitions_u19_df['frac_diss']*lwtransitions_u19_df['f'],
        bins=100, range=(wl_min, wl_max), histtype='step', log=True)
ax.hist(lwtransitions_abgrall_df['wl'],
        weights=lwtransitions_abgrall_df['frac_diss']*lwtransitions_abgrall_df['f'],
        bins=100, range=(wl_min, wl_max), histtype='step', log=True)
```
